```python
%reload_ext autoreload
%autoreload 2
import pandas as pd
import numpy as np 
```

```python
train_df = pd.read_csv("data/bank/train.csv")
test_df = pd.read_csv("data/bank/test.csv")

X_train = train_df[train_df.columns[:-1]]
y_train = train_df[train_df.columns[-1]]

X_test = test_df[test_df.columns[:-1]]
y_test = test_df[test_df.columns[-1]]

train_df.head()
```

```python
from splearn.EnsembleLearning.RandomForest import RandomForest
from splearn.DecisionTree.DecisionTree import DecisionTree
import splearn.Metrics as spmet

columns = [
    "num_trees",
    "train_acc",
    "test_acc"
]

num_obs = 10
points = [2 ** i for i in range(num_obs)]

rf = RandomForest()
rf.train(
    X_train,
    y_train,
    0,
    seed = 42,
    gain = "gini"
)

data = []

for i in points:

    while len(rf) < i:
        rf.iterate(1)
        print(f"Iteration {len(rf)} complete!")

    p_train = rf.predict(X_train)
    p_test  = rf.predict(X_test)

    data.append([
        i,
        spmet.accuracy_score(p_train, y_train),
        spmet.accuracy_score(p_test, y_test),
    ])

    print('\t', "Observation complete!")

data = np.array(data)
df = pd.DataFrame(data, columns = columns)
df.to_csv("randomforest.csv")

print("Experiment complete!")
```

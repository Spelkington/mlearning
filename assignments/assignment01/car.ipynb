```python
from ID3Tree import ID3Tree
import numpy as np
import pandas as pd
```

```python
car_df = pd.read_csv("car.csv")
car_df
```

```python
from sklearn.model_selection import train_test_split
car_train, car_test = train_test_split(
    car_df,
    test_size = 0.2,
    train_size = 0.8,
)

train_feat = car_train[car_train.columns[:-1]]
train_targ = car_train[car_train.columns[-1]]

test_feat = car_test[car_test.columns[:-1]]
test_targ = car_test[car_test.columns[-1]]

from sklearn.metrics import accuracy_score

types = [
    "entropy",
    "majority error",
    "gini"
]

obs = 10

results = np.zeros((obs, len(types)))
results = pd.DataFrame(
    results,
    columns = types
)

for t in types:
    
    print(f"Running {t}...")
    
    data = np.zeros(len(results))

    for i in range(len(results)):
        
        depth = i+1
        tree = ID3Tree()
        tree.train(train_feat,
                   train_targ,
                   t,
                   max_depth = depth
        )
        trained_targ = tree.predict(test_feat)
        
        data[i] = accuracy_score(trained_targ, test_targ)
        
    results[t] = data

results
```

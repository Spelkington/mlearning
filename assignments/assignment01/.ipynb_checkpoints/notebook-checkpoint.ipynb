```python
%reload_ext autoreload
%autoreload 2
from splearn.DecisionTree import DecisionTree
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data/car/car.csv")
df
```

```python
features = df[df.columns[:-1]]
target   = df[df.columns[-1]]
```

```python
from sklearn.metrics import accuracy_score
from splearn.DecisionTree import DecisionTree

dt = DecisionTree()
dt.train(features, target, gain="entropy")
preds = dt.predict(features)

print(accuracy_score(target, preds))
print(len(dt.dump_nodes()))
```

```python
dt2 = DecisionTree()
dt2.train(features, target, gain="entropy", max_depth=1)
preds = dt2.predict(features)

print(accuracy_score(target, preds))
print(len(dt2.dump_nodes()))
```

```python
bank_df = pd.read_csv("data/bank/train.csv")
bank_df.head()
```

```python
bank_targ = bank_df[bank_df.columns[-1]]
bank_feat = bank_df[bank_df.columns[:-1]]
```

```python
from splearn.DecisionTree import DecisionTree

bank_dt = DecisionTree()
bank_dt.train(
    bank_feat,
    bank_targ, 
    gain="entropy",
)

preds = bank_dt.predict(bank_feat)

print(accuracy_score(bank_targ, preds))
print(len(bank_dt.dump_nodes()))
```

```python
from splearn.DecisionTree import DecisionTree

weights = pd.Series([1 + (i % 2) for i in range(len(bank_targ))])

bank_wdt = DecisionTree()
bank_wdt.train(
    bank_feat,
    bank_targ,
    weights,
    gain="entropy"
)

preds = bank_wdt.predict(bank_feat)

print(accuracy_score(bank_targ, preds))
print(len(bank_wdt.dump_nodes()))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(bank_targ, preds)
```

```python
from sklearn.model_selection import train_test_split
from splearn.DecisionTree import DecisionTree

X_train, X_test, y_train, y_test = train_test_split(
    bank_feat,
    bank_targ, 
    test_size=0.25
)

depths = 16
types = [
    "entropy",
    "majority error",
    "gini"
]

data = np.zeros((depths, len(types)))

for i, d in enumerate(range(depths)):
    for j, t in enumerate(types):
        
        print(d, t)

        bank_dt = DecisionTree()
        bank_dt.train(
            X_train,
            y_train,
            gain=t,
            max_depth=d
        )

        y_pred = bank_dt.predict(X_test)
        
        data[i,j] = accuracy_score(y_test, y_pred)
```

```python
df = pd.DataFrame(data, columns = types)
df
```
